```python
import sqlite3
from sqlite3 import Error 
import pandas as pd 
```

```python
def create_connection(db_path=None): 
    try:
        conn = sqlite3.connect(db_path)
    except:
        raise 
    else:
        return conn
    
```

```python
def execute_sql(conn, sql_command):
    try:
        cursor = conn.cursor()
        cursor.execute(sql_command)
        res = cursor.fetchall()
    except:
        raise 
    else:
        return res
   
    
```

```python
conn = create_connection("sample.db")
```

```python
res = execute_sql(conn, """
CREATE TABLE IF NOT EXISTS genes ([gene_id] TEXT PRIMARK KEY,
[gene_expression] NUMERIC)
""")
```

```python
def insert_values(values, table, conn):
    return execute_sql(conn, f"INSERT OR REPLACE INTO {table} {values}")
```
